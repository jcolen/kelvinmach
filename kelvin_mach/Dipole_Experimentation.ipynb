```python
import wake_generalized_source as wgs
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = 1.4   #Alternatively 4
l = 0.05  #0.09 for M = 4
R = 256
d = 0.125
```

```python
phi = wgs.generate_wake_grid(M, 1.0, 1.0, l, R, d, wgs.delta)
```

```python
dipole_separation = 40
dsep_pixels = int(dipole_separation / d)
```

```python
#Vertically separated dipole
phi1 = np.pad(phi, ((dsep_pixels, 0), (0, 0)), 'constant', constant_values=((0, 0), (0, 0)))
phi2 = np.pad(-phi, ((0, dsep_pixels), (0, 0)), 'constant', constant_values=((0, 0), (0, 0)))
phi_total = phi1 + phi2
plt.imshow(np.real(phi_total))
plt.show()
```

```python
#Horizontally separated dipole
phi1 = np.pad(phi, ((0, 0), (dsep_pixels, 0)), 'constant', constant_values=((0, 0), (0, 0)))
phi2 = np.pad(-phi, ((0, 0), (0, dsep_pixels)), 'constant', constant_values=((0, 0), (0, 0)))
phi_total = phi1 + phi2
plt.imshow(np.real(phi_total))
plt.show()
```

```python
#Quadrupole Diagonally Oriented
d2 = dsep_pixels / 2
phi1 = np.pad(phi, ((d2, d2), (dsep_pixels, 0)), 'constant', constant_values=((0, 0), (0, 0)))
phi2 = np.pad(phi, ((d2, d2), (0, dsep_pixels)), 'constant', constant_values=((0, 0), (0, 0)))
phi3 = np.pad(-phi, ((dsep_pixels, 0), (d2, d2)), 'constant', constant_values=((0, 0), (0, 0)))
phi4 = np.pad(-phi, ((0, dsep_pixels), (d2, d2)), 'constant', constant_values=((0, 0), (0, 0)))
phi_total = phi1 + phi2 + phi3 + phi4
plt.imshow(np.real(phi_total))
plt.show()
```

```python
#Quadrupole Rectangularly Oriented
phi1 = np.pad(phi, ((dsep_pixels, 0), (dsep_pixels, 0)), 'constant', constant_values=((0, 0), (0, 0)))
phi2 = np.pad(phi, ((0, dsep_pixels), (0, dsep_pixels)), 'constant', constant_values=((0, 0), (0, 0)))
phi3 = np.pad(-phi, ((dsep_pixels, 0), (0, dsep_pixels)), 'constant', constant_values=((0, 0), (0, 0)))
phi4 = np.pad(-phi, ((0, dsep_pixels), (dsep_pixels, 0)), 'constant', constant_values=((0, 0), (0, 0)))
phi_total = phi1 + phi2 + phi3 + phi4
plt.imshow(np.real(phi_total))
plt.show()
```
